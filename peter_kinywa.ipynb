```python
# Use a variable 'features' to capture all the features (independent variables)
features = ['Rooms','Distance','Bedroom2','Bathroom','BuildingArea','Car','CouncilArea','Lattitude','Longtitude']
X = data[features] # Assigns all the features to the variable 'X'
y = data['Price'] # Assigns the target variable to 'y'
```

```python
# Split the data into the training set and the validation set.
# Set the training set to 70 percent of the observations. The splitting is done randomly.
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=0)
```

```python
# View the X_Train dataframe of the training dataset 
X_train
```

```python
# Training the model using linear regression
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
model=regressor.fit(X_train, y_train) # Generates (fit) the model.
```

```python
# Generate the intercept
model.intercept_
```

```python
# Generate the coefficients
coef=pd.DataFrame ({'Features':features,'Coefficients':model.coef_})
```

# Polynomial Regression With scikit-learn

```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
features = ['Rooms','Distance','Bedroom2','Bathroom','BuildingArea','Car','CouncilArea','Lattitude','Longtitude']
X = data[features] 
y = data['Price']
X, y = np.array(X), np.array(y)
```

```python
transformer = PolynomialFeatures(degree=2, include_bias=False)
transformer.fit(X)
```

```python
X_ = transformer.transform(X)
X_ = PolynomialFeatures(degree=2, include_bias=False).fit_transform(X)
model = LinearRegression().fit(X_, y)
```

```python
 r_sq = model.score(X_, y)
print('coefficient of determination:', r_sq)
print('intercept:', model.intercept_)
print('coefficients:', model.coef_)
```
